```python
%matplotlib inline
```

```python
# Import dependencies

import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# creating the plot

plt.plot(x_axis, y_axis)
plt.show()
```

```python
# create the plot with ax.plt()

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()

ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

ax = plt.axes()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend using plt.

plt.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")

plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

# Create a title.

plt.title("PyBer Fare by Month")

# add a grid

plt.grid()

# Add the legend.

plt.legend()
```

```python
# creating a bar plot using plt.

plt.bar(x_axis, y_axis, color="green", label='Boston')

# create lables for the axis

plt.xlabel("Date")
plt.ylabel("Fare($)")

# create a title

plt.title("PyBer Fare by Month")

# add a legend

plt.legend
```

```python
# creating a horizontal bar plot using plt.

plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()

# create lables for the axis

plt.xlabel("Fare($)")
plt.ylabel("Date")

# create a title

plt.title("PyBer Fare by Month")

# add a legend

plt.legend
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()

ax.invert_yaxis()

# create horizontal bar graph and color is cyan

ax.barh(x_axis, y_axis, color="cyan", label="Chicago")

# add labels to the axis

ax.set_xlabel('Fare($)')
ax.set_ylabel('Month')

# add a legend

ax.legend()
```

```python
# creating a scatter plot using plt.plot()

plt.plot(x_axis, y_axis, 'o')
```

```python
# creating a scatter plot using plt.scatter()

plt.scatter(y_axis, x_axis, c='red', label="Chicago")

plt.gca().invert_yaxis()

# adding axis labels and title

plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.title("PyBer Fare by Month")

#adding a limit to the x axis

plt.xlim(0, 50)

plt.legend()
```

```python
# creating a bubble plot using plt.scatter() 

# create a for loop to multiply fares to make bubbles larger

#y_axis_larger = []
#for data in y_axis:
    #y_axis_larger.append(data*3)

# using list comprehension to create larger bubbles    
    
plt.scatter(y_axis, x_axis, s = [i * 3 for i in y_axis], c='red', label="Chicago")

plt.gca().invert_yaxis()

# adding axis labels and title

plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.title("PyBer Fare by Month")

#adding a limit to the x axis

plt.xlim(0, 50)
```

```python
# create a scatter plot using ax.scatter()

fig, ax = plt.subplots()

ax.scatter(x_axis, y_axis)
```

```python
# creating a bubble plot using ax.scatter

fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, c="lightblue", linewidths=2, alpha=.8, s = [i * 5 for i in y_axis], label="Boston")

plt.gca().invert_yaxis()

# adding axis label and title

ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.set_title("PyBer Fare by Month")

# adding min and max to x axis

ax.set_xlim(0, 50)

# adding a legend

ax.legend()
```

```python
# creating a pie chart with plt.pie()

# assigning colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# choosing slice to explode
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# increasing the size of the figure
plt.subplots(figsize=(8, 8))

# parameters for chart
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```

```python
# creating a pie chart using ax.pie()

fig, ax = plt.subplots(figsize=(8, 8))

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values2 = [0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]

# for object-oriented you need to specify size in initial creating, not later on like MATLAB model
#fig, ax = plt.subplots(figsize=(8, 8))

ax.pie(y_axis, 
       startangle=95,
       explode=explode_values2,
       colors=colors,
       shadow=True,
       labels=x_axis, 
       counterclock=False,
       autopct='%.1f%%')

plt.show()
```
